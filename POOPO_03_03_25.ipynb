```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el radio del círculo
r = 1

# Crear un array de ángulos theta de 0 a 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

# Calcular las coordenadas x e y usando las ecuaciones paramétricas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Crear la gráfica
plt.figure(figsize=(6, 6))
plt.plot(x, y)

# Añadir etiquetas y título
plt.title('Círculo usando ecuaciones paramétricas')
plt.xlabel('x')
plt.ylabel('y')

# Ajustar el aspecto para que el círculo no se vea elíptico
plt.gca().set_aspect('equal', adjustable='box')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros de la hélice
r = 1  # Radio de la hélice
c = 0.2  # Constante que controla la altura por vuelta

# Crear un array de valores para el parámetro t
t = np.linspace(0, 8 * np.pi, 1000)  # 4 vueltas completas

# Ecuaciones paramétricas de la hélice
x = r * np.cos(t)
y = r * np.sin(t)
z = c * t

# Crear la gráfica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la hélice
ax.plot(x, y, z, label='Hélice', color='blue')

# Añadir etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Hélice usando ecuaciones paramétricas')

# Añadir una leyenda
ax.legend()

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un array de ángulos theta de 0 a 2*pi
theta = np.linspace(0, 2 * np.pi, 1000)

# Ecuaciones paramétricas para el corazón
x = 16 * np.sin(theta)**3
y = 13 * np.cos(theta) - 5 * np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

# Crear la gráfica
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red')

# Añadir etiquetas y título
plt.title('Corazón usando ecuaciones paramétricas')
plt.xlabel('x')
plt.ylabel('y')

# Ajustar el aspecto para que el corazón se vea correctamente
plt.gca().set_aspect('equal', adjustable='box')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(6, 8))

# Cuerpo principal (rectángulo con bordes redondeados)
body = patches.FancyBboxPatch((0.2, 0.1), 0.6, 0.8, boxstyle="round,pad=0.1",
                              edgecolor="black", facecolor="red", linewidth=2)
ax.add_patch(body)

# Mochila (rectángulo más pequeño)
backpack = patches.Rectangle((0.1, 0.3), 0.1, 0.5, edgecolor="black", facecolor="red", linewidth=2)
ax.add_patch(backpack)

# Visor (círculo)
visor = patches.Circle((0.5, 0.7), 0.15, edgecolor="black", facecolor="cyan", linewidth=2)
ax.add_patch(visor)

# Piernas (rectángulos)
leg1 = patches.Rectangle((0.3, 0.05), 0.2, 0.1, edgecolor="black", facecolor="red", linewidth=2)
leg2 = patches.Rectangle((0.5, 0.05), 0.2, 0.1, edgecolor="black", facecolor="red", linewidth=2)
ax.add_patch(leg1)
ax.add_patch(leg2)

# Ajustar los límites del gráfico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Ocultar los ejes
ax.axis("off")

# Mostrar la figura
plt.show()
```
